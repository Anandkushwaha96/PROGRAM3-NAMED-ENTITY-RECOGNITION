```python
import numpy as np

# Define the data (input, expected output)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])  # Output for AND gate

# Define weights and biases (initialize randomly)
weights1 = np.random.rand(2, 3)  # 2 input neurons, 3 hidden layer neurons
bias1 = np.random.rand(3)

weights2 = np.random.rand(3, 1)  # 3 hidden neurons, 1 output neuron
bias2 = np.random.rand(1)

# Define learning rate
learning_rate = 0.1

# Activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Training for 1 epoch (you can use a loop for multiple epochs)
for i in range(len(X)):
    x = X[i]
    target = y[i]

    # Forward pass
    z1 = np.dot(x, weights1) + bias1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, weights2) + bias2
    a2 = sigmoid(z2)

    output = a2
    error = target - output

    # Backpropagation
    output_derivative = sigmoid_derivative(z2)
    hidden2_derivative = output_derivative * error

    weights2_derivative = np.dot(a1.reshape(-1, 1), hidden2_derivative.reshape(1, -1))
    weights2 -= learning_rate * weights2_derivative
    bias2 -= learning_rate * hidden2_derivative

    hidden1_derivative = np.dot(weights2, hidden2_derivative) * sigmoid_derivative(z1)
    weights1_derivative = np.dot(x.reshape(-1, 1), hidden1_derivative.reshape(1, -1))
    weights1 -= learning_rate * weights1_derivative
    bias1 -= learning_rate * hidden1_derivative

# Print the final weights and biases
print("Updated Weights (Hidden Layer 1):", weights1)
print("Updated Bias (Hidden Layer 1):", bias1)
print("Updated Weights (Hidden Layer 2):", weights2)
print("Updated Bias (Hidden Layer 2):", bias2)
```
